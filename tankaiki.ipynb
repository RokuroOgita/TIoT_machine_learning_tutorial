```python
def show_graph(X, Y, x, y):
    fig, ax = plt.subplots(dpi=100)
    ax.scatter(X, Y, marker='.')
    ax.plot(x, y, 'tab:red')
    ax.set_title('最高気温とアイスクリーム・シャーベットの支出額')
    ax.set_xlabel('最高気温の月平均（℃）')
    ax.set_ylabel('支出額（円）')
    ax.set_xlim(0, 35)
    ax.set_ylim(-250, 2000)
    ax.grid()
    fig.show()
    
```

```python
W = np.polyfit(X, Y, 1)# 単回帰の場合3つめの引数に1を指定。線形回帰モデルのパラメータを吐き出す。
W
```

```python
x = np.linspace(0, 35, 100)
y_hat = np.polyval(W, x)# 目的変数の推定
show_graph(X, Y, x, y_hat)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X.reshape(-1, 1), Y)
LinearRegression()
```

```python
# 回帰直線の係数(傾き)
reg.coef_
```

```python
# 回帰直線の切片
reg.intercept_
```

```python
# 決定係数を求める
reg.score(X.reshape(-1, 1), Y)
```

```python
x = np.linspace(0, 35, 100)
y_hat = reg.predict(x.reshape(-1, 1))# 目的変数
show_graph(X, Y, x, y_hat)
```

```python
# 確認問題
D = np.array([[1, 3], [3, 6], [6, 5], [8, 7]])
X = D[:,0]
Y = D[:,1]
```

```python
reg = LinearRegression()
reg.fit(X.reshape(-1, 1), Y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(X.reshape(-1, 1), Y)# 決定係数(R^2) / 0.5を超えたらそこそこ良いとされる。
# from sklearn.metrics import r2_score にも似たような計算モジュールが有る。
```

```python
y_hat = reg.predict(x.reshape(-1, 1))
def show_graph(X, Y, x, y):
    fig, ax = plt.subplots(dpi=100)
    ax.scatter(X, Y, marker='.')
    ax.plot(x, y, 'tab:red')
    ax.set_title('X')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.grid()
    fig.show()
show_graph(X, Y, x, y_hat)
```
