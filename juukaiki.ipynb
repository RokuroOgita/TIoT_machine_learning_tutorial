```python
# https://chokkan.github.io/mlnote/regression/02mra.html

import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
a, b = 0.431, 3.310 
D = np.array([[1, 3], [3, 6], [6, 5], [8, 7]])

fig, ax = plt.subplots(dpi=100, figsize=(6, 6))

ax.scatter(D[:,0], D[:,1], marker='o')

x = np.array([0, 10])
ax.plot(x, a * x + b, 'tab:red', ls='-', label=f'${a}x + {b}$')

for i, row in enumerate(D):
    x, y = row
    y_hat = a * x + b
    plt.vlines([x], min(y, y_hat), max(y, y_hat), "tab:blue", linestyles='dashed')

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_aspect('equal')
ax.grid()
plt.legend(loc='upper left')
plt.show()
```

```python
# sklearn.linear_model.LinearRegressionを用いる例
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
reg = make_pipeline(PolynomialFeatures(3), LinearRegression())# 三次関数にフィットさせる例
reg.fit(X.reshape(-1, 1), Y)
```

```python
reg[1].coef_
```

```python
reg[1].intercept_
```

```python
reg.score(X.reshape(-1, 1), Y)# 決定係数(R^2)
```

```python
x = np.linspace(0, 35, 100)
y_hat = reg.predict(x.reshape(-1, 1))
plot_graph(X, Y, x, y_hat)
```

```python
# 確認問題

D = np.array([[1, 3], [3, 6], [6, 5], [8, 7]])
X = D[:, 0]
Y = D[:, 1]
def plot_graph(X, Y, x, y):
    fig, ax = plt.subplots(dpi=100)
    ax.set_title('X')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.grid()
    ax.scatter(X, Y, marker='.')
    ax.plot(x, y, 'r')
    fig.show()
    
reg = make_pipeline(PolynomialFeatures(3), LinearRegression())# 三次関数にフィットさせる例
reg.fit(X.reshape(-1, 1), Y)
```

```python
x = np.linspace(0, 10, 100)
y_hat = reg.predict(x.reshape(-1, 1))
plot_graph(X, Y, x, y_hat)
```

```python
reg.score(X.reshape(-1, 1), Y)# 決定係数
reg[1].coef_
```
